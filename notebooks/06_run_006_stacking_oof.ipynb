# 006 - Stacking (Out-Of-Fold)

## Objectif
Utiliser le **Stacking** pour combiner plusieurs modèles. Contrairement au Blending simple, le Stacking utilise un méta-modèle (ici Ridge) pour apprendre la meilleure combinaison des prédictions des modèles de base.

## Méthodologie (OOF)
Pour éviter le data leakage, on utilise des prédictions **Out-Of-Fold (OOF)** :
1. On divise le train set en K folds.
2. Pour chaque fold, on entraîne les modèles de base sur (K-1) folds et on prédit sur le fold restant.
3. On obtient ainsi une prédiction pour chaque point du train set, faite par un modèle qui n'a pas vu ce point pendant l'entraînement.
4. Ces prédictions OOF servent de features pour entraîner le méta-modèle.

### Modèles de Base
- **XGBoost**: Tuned + Feature Engineering
- **CatBoost**: Native Categorical + Log Target
- **ExtraTrees**: Randomization for diversity

### Méta-Modèle
- **Ridge Regression**: Régression linéaire régularisée (L2).
